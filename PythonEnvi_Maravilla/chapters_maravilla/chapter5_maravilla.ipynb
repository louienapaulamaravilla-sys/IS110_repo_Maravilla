Exercise 1

Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```

1. == (Equality Operator) – checks if two values are equal.

2. is (Identity Operator) – checks if two variables refer to the same object in memory.

```python
Comparisons
1. a == b → True
2 and 2.0 have the same value, even though one is an integer and the other is a float.

2. a is b → False
They are different objects in memory because their data types are different (int vs float).

3. a == c → True
c was assigned the value of a, so their values are equal.

4. a is c → True
c refers to the same object in memory as a.
```

Challenge 1

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333

````python
total = 0
count = 0

while True:
    num = input("Enter a number: ")

    if num == "done":
        break

    try:
        value = float(num)
        total = total + value
        count = count + 1
    except:
        print("Invalid input")
        continue

if count > 0:
    average = total / count
    print(total, count, average)
````

Challenge 2

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

````python
max_num = None
min_num = None

while True:
    num = input("Enter a number: ")

    if num == "done":
        break

    try:
        value = float(num)
    except:
        print("Invalid input")
        continue

    if max_num is None or value > max_num:
        max_num = value

    if min_num is None or value < min_num:
        min_num = value

print("Maximum:", max_num)
print("Minimum:", min_num)
````
