A. Create an ndarry object (A) with any values from 1 to 100 in a nested Python List. You can choose your range from the given values (1-100).

* Given that A is an ndarray object, many functionalities in numpy could be accessed by:
    + Make sure to explain each line of codes that you have created.

1. Using a property of the ndarray object
      - e.g. `A.shape`
2. Using a method of the ndarray object
      - e.g. `A.max()`
3. Using a numpy function and passing an ndarray
     - e.g. `np.max(A)`

```python
import numpy as np

# Create a nested list
data = [
    [10, 20, 25, 30, 40],
    [45, 50, 60, 65, 70],
    [75, 80, 90, 95, 100]
]

# Convert the nested list into a NumPy array
A = np.array(data)

# Display the ndarray
print("Array A:")
print(A)


# 1. Using a property of ndarray
# A.shape shows the size of the array (rows, columns)
print("\n1. Property: A.shape")
print("Shape of A:", A.shape)


# 2. Using a method of ndarray
# A.max() returns the highest value in the array
print("\n2. Method: A.max()")
print("Maximum value in A:", A.max())


# 3. Using a numpy function
# np.max(A) also returns the highest value
print("\n3. NumPy Function: np.max(A)")
print("Maximum value using NumPy:", np.max(A))
```
