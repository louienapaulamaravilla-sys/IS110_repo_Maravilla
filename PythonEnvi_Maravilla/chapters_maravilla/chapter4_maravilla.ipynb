Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0

```python
def comp_pay(hours, rate):
    if hours > 40:
        overtime_hours = hours - 40
        pay = (40 * rate) + (overtime_hours * rate * 1.5)
    else:
        pay = hours * rate
    return pay


input_hours = input("Enter Hours: ")
hours = float(input_hours)
input_rate = input("Enter Rate: ")
rate = float(input_rate)

pay = comp_pay(hours, rate)
print("Pay:", pay)
```

CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		1.75
    [ 82,  85)		2.0
    [ 78,  81)		2.25
    [ 75,  77)		2.5
    [ 0,   74)		3.0

```python
def computegrade(score):
    if score >= 98 and score <= 100:
        return "1.0"
    elif score >= 94:
        return "1.25"
    elif score >= 90:
        return "1.5"
    elif score >= 86:
        return "1.75"
    elif score >= 82:
        return "2.0"
    elif score >= 78:
        return "2.25"
    elif score >= 75:
        return "2.5"
    elif score >= 0:
        return "5.0"
    else:
        return "Invalid score"

input_score = input("Enter Score: ")
score = float(input_score)
grade = computegrade(score)
print("Grade:", grade)
```
